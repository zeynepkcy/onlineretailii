# **Feature Engineering**

```python
# Converting InvoiceDate to datetime. InvoiceDate is in format of 01-12-2010 08:26:33.
retail_df["InvoiceDate"] = pd.to_datetime(retail_df["InvoiceDate"], format="%Y-%m-%d %H:%M:%S")
```

```python
retail_df["year"] = retail_df["InvoiceDate"].apply(lambda x: x.year)
retail_df["month_num"] = retail_df["InvoiceDate"].apply(lambda x: x.month)
retail_df["day_num"] = retail_df["InvoiceDate"].apply(lambda x: x.day)
retail_df["hour"] = retail_df["InvoiceDate"].apply(lambda x: x.hour)
retail_df["minute"] = retail_df["InvoiceDate"].apply(lambda x: x.minute)
retail_df["second"] = retail_df["InvoiceDate"].apply(lambda x: x.second)
```

```python
# extracting day from the Invoice date
retail_df['Day']=retail_df['InvoiceDate'].dt.day_name()
```

Miktar ile fiyatın çarpılmasıyla toplam tutar sütununun oluşturulması

```python
retail_df['TotalAmount']=retail_df['Quantity']*retail_df['Price']
```

TOP 10 HIGHEST SELLING PRODUCTS SOLD BY THE STORE

```python
df1=retail_df.groupby('Description').sum()
df1.sort_values(['Quantity'], ascending=False,inplace=True)
df1.reset_index(inplace=True)
df1.rename(columns={'Description':'Product_name'},inplace=True)
df2=df1[['Product_name','Quantity']][:10]
df2
```

```python
# top 10 products by quantity
plt.figure(figsize=(12,6))
sns.barplot(x=df2['Quantity'],y=df2['Product_name'])
plt.title('Top 10 products by quantity')
```

10 LEAST SELLING PRODUCTS OF THE STORE

```python
df3=df1[['Product_name','Quantity']].tail(10)
df3
```

MİKTAR BAZINDA İLK 10 STOK KODU

```python
df4=retail_df.groupby('StockCode').sum()
df4.sort_values(['Quantity'], ascending=False,inplace=True)
df4.reset_index(inplace=True)
df5=df4[['StockCode','Quantity']][:10]
df5
```

```python
# top 10 StockCodes by quantity
plt.figure(figsize=(12,6))
sns.barplot(x=df5['Quantity'],y=df5['StockCode'])
plt.title('Top 10 StockCodes by quantity')
```

```python
import pandas as pd

# retail_df DataFrame'inizi burada yükleyin veya oluşturun. Örneğin:
# retail_df = pd.read_csv('your_data_file.csv')

# 'InvoiceDate' sütununu datetime formatına dönüştürme
retail_df["InvoiceDate"] = pd.to_datetime(retail_df["InvoiceDate"], format="%d-%m-%Y %H:%M:%S")

# Yıl ve ay bilgilerini çıkararak yeni sütunlar oluşturma
retail_df["Year"] = retail_df["InvoiceDate"].dt.year
retail_df["Month"] = retail_df["InvoiceDate"].dt.month_name()

# 'Month' ve 'TotalAmount' kullanarak aylık toplam satışları hesaplama ve sıralama
sales_by_month = retail_df.groupby('Month')['TotalAmount'].sum().reset_index()

# Ayları ayın sırasına göre sıralama için bir sıralama anahtarı oluşturma
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
                'August', 'September', 'October', 'November', 'December']
sales_by_month['Month'] = pd.Categorical(sales_by_month['Month'], categories=months_order, ordered=True)

# Sonuçları azalan sırada sıralama
sales_by_month_sorted = sales_by_month.sort_values('TotalAmount', ascending=False)

# Sonuçları görüntüleme
print(sales_by_month_sorted)
```

```python
# Sales different months.
plt.figure(figsize=(20,6))
sns.barplot(x=sales_by_month['Month'],y=sales_by_month['TotalAmount'])
plt.title('Sales in different Months ')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming the DataFrame 'retail_df' is already loaded with the necessary data.
# The 'month_num' and 'TotalAmount' columns should be present for this operation.

# Ensure month names are in the correct order for plotting
month_order = ["January", "February", "March", "April", "May", "June", "July",
               "August", "September", "October", "November", "December"]

# Group by month and sum the total amounts
sales_by_month = retail_df.groupby(retail_df['InvoiceDate'].dt.strftime('%B'))['TotalAmount'].sum().reindex(month_order)

# Plotting the results
plt.figure(figsize=(10, 6))
sns.barplot(x=sales_by_month.index, y=sales_by_month.values, palette='viridis')
plt.xticks(rotation=45)
plt.title('Total Sales by Month')
plt.ylabel('Total Amount')
plt.xlabel('Month')
plt.tight_layout()
plt.show()

# Please note that the plotting section should be run in a Jupyter notebook or a similar environment that supports plotting.
# Also, the seaborn library should be installed in the environment where this code is run. If it is not installed, you can install it using pip:
# !pip install seaborn

# The code is commented out to prevent execution in this environment. To execute, please run this in an appropriate Python environment.
```
